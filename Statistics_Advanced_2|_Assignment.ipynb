Question 1: What is hypothesis testing in statistics?
-->
    Hypothesis testing in statistics is a process for deciding whether a sample provides enough evidence to reject a claim about the population. It makes null hypothesis and alternate hypothesis, collects the sample data. Using this sample data evidence it rejects or accepts the claim. It helps to the researchers to write decision.
    Null Hypothesis: The initial assumption or claim that there is no effect, no difference, or nothing interesting happening.
    Alternate Hypothesis: The opposite of the null hypothesis, which suggests that there is an effect or difference.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
-->
    The null hypothesis is indication of there is no relation or effect between two variables or data. While alternate hypothesis states that there is relation between that variables which is opposite of the Null Hypothesis. The null hypothesis (H₀) is the default assumption that there is no effect, no difference, or no relationship between variables.


Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
-->
    The significance level (Alpha) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true it is type1 error. If the p-value from the test is less than or equal to the significance level the null hypothesis is rejected otherwise it is accepted.


Question 4: What are Type I and Type II errors? Give examples of each.
-->
    Type I Error: Type I error occurs when true null hypothesis is rejected wrongly.
    Example: Sending an innocent person to jail for a crime they didn’t commit.
    Type II Error: It occurs when false null hypothesis is not rejected.
    Example: Concluding that a new medicine doesn’t work when it actually does.

```python
''' Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
 '''

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy import stats

population_mean=np.mean(sample_data)
population_std=np.std(sample_data)
alpha=0.05
n=len(sample_data)

zstat=(population_mean-50)/(population_std/np.sqrt(n))
print("Z_statistics:",zstat)
p_value=2*(1-stats.norm.cdf(abs(zstat)))
print("P_value is:",p_value)

if p_value<alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
''' Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. '''

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data=np.random.normal(99, 20, 20)
sample_mean=np.mean(data)
sample_std=np.std(data)
n=len(data)
alpha=0.05
z_critical=stats.norm.ppf(1-alpha/2)
margin_of_error=z_critical*(sample_std/np.sqrt(n))
confidence_interval=(sample_mean-margin_of_error, sample_mean+margin_of_error)
print("sample mean",sample_mean)
print("sample std",sample_std)
print("Confidence Interval:",confidence_interval)

plt.hist(data,bins=10,color="red", edgecolor='black')
plt.title("Normal Distribution with 95%  confidence interval")
plt.axvline(sample_mean, color='black', linestyle='--')
plt.axvline(confidence_interval[0], color='green', linestyle='--',label="Lower confidence")
plt.axvline(confidence_interval[1], color='green', linestyle='--',label="Upper confidence")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
''' Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
 '''

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean=np.mean(data)
    std=np.std(data)
    z_scores=(data-mean)/std
    return z_scores

z_scores=calculate_z_scores(np.random.normal(0, 1, 100))

plt.hist(z_scores, bins=10, edgecolor='black')
plt.title("Standardized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
